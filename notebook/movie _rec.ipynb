```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("C:\\Users\\ADMIN\\Downloads\\movie_rec_dataset\\movies_metadata.csv", low_memory=False)
movies.head()
```

```python
movies = movies[['id', 'title', 'genres', 'overview']]
movies.head()
```

```python
movies = movies[movies['id'].str.isnumeric()]
movies['id'] = movies['id'].astype(int)
```

```python
credits = pd.read_csv("C:\\Users\\ADMIN\\Downloads\\movie_rec_dataset\\credits.csv")
credits.head()
```

```python
keywords = pd.read_csv("C:\\Users\\ADMIN\\Downloads\\movie_rec_dataset\\keywords.csv")
keywords.head()
```

```python
movies = movies.merge(credits, on='id')
movies = movies.merge(keywords, on='id')
```

```python
movies.head()
movies.shape
movies.info()
```

```python
movies['overview'] = movies['overview'].fillna('')
movies['genres'] = movies['genres'].fillna('[]')
movies['keywords'] = movies['keywords'].fillna('[]')
movies['cast'] = movies['cast'].fillna('[]')
movies['crew'] = movies['crew'].fillna('[]')
```

```python
import ast

def convert(text):
    try:
        return [i['name'] for i in ast.literal_eval(text)]
    except:
        return []

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert_cast(text):
    try:
        return [i['name'] for i in ast.literal_eval(text)[:3]]
    except:
        return []

movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
def fetch_director(text):
    try:
        for i in ast.literal_eval(text):
            if i['job'] == 'Director':
                return [i['name']]
    except:
        return []
    return []

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    return [i.replace(" ", "") for i in L]
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
```

```python
movies['tags'] = (
    movies['overview'] + ' ' +
    movies['genres'].apply(lambda x: " ".join(x)) + ' ' +
    movies['keywords'].apply(lambda x: " ".join(x)) + ' ' +
    movies['cast'].apply(lambda x: " ".join(x)) + ' ' +
    movies['crew'].apply(lambda x: " ".join(x))
)
```

```python
movies['tags'] = movies['tags'].str.lower()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    return " ".join([ps.stem(word) for word in text.split()])

movies['tags'] = movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

movies = movies.drop_duplicates(subset='title')
movies.reset_index(drop=True, inplace=True)

tfidf = TfidfVectorizer(max_features=5000,stop_words='english')
tfidf_vectors = tfidf.fit_transform(movies['tags']).toarray()  # sparse matrix

from sentence_transformers import SentenceTransformer

semantic_model = SentenceTransformer('all-MiniLM-L6-v2')

semantic_vectors = semantic_model.encode(
    movies['tags'].tolist(),
    show_progress_bar=True
)
```

```python
def get_recommendations(movie_title, tfidf_vectors, semantic_vectors, movies_df, top_n=10):

    titles = movies_df['title'].fillna('').str.lower()

    # Exact match first
    matches = movies_df[titles == movie_title.lower()]

    # Fallback partial match
    if matches.empty:
        matches = movies_df[titles.str.contains(movie_title.lower(), na=False)]

    if matches.empty:
        return "Movie not found"

    idx = matches.index[0]

    from sklearn.metrics.pairwise import cosine_similarity

    tfidf_scores = cosine_similarity(
        tfidf_vectors[idx].reshape(1, -1),
        tfidf_vectors
    ).flatten()

    semantic_scores = cosine_similarity(
        semantic_vectors[idx].reshape(1, -1),
        semantic_vectors
    ).flatten()

    final_scores = (0.5 * tfidf_scores) + (0.5 * semantic_scores)

    similar_movies = sorted(
        list(enumerate(final_scores)),
        key=lambda x: x[1],
        reverse=True
    )

    seen_titles = set()
    recommendations = []

    for i, score in similar_movies:
        title = movies_df.iloc[i].title

        if title.lower() != movie_title.lower() and title.lower() not in seen_titles:
            seen_titles.add(title.lower())
            recommendations.append(title)

        if len(recommendations) == top_n:
            break

    return recommendations
```

```python
get_recommendations("shark", tfidf_vectors, semantic_vectors, movies)
```
